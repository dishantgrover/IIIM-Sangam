```python
def normalize(df,columns):
    df1 = df.copy()
    for column in columns:
        df1[column] = (df[column]-df[column].mean())/df[column].std()
    return df1
```

```python
cont_cols.remove('date_time')
```

```python
total_norm = normalize(total,cont_cols)
```

```python
import math
```

```python
means = []
stds = []
for column in cont_cols:
    m = total_norm[column].mean()
    s = total_norm[column].std()
    means.append(m);stds.append(s)
    assert m<1e-3
    assert math.fabs(s-1)<1e-3
```
