```python
from sklearn.preprocessing import MinMaxScaler
def normalize(train,test,label):
    sc = MinMaxScaler()
    sc.fit(pd.concat([train[label],test[label]],0).values.reshape(-1,1))
    return sc.transform(train[label].values.reshape(-1,1)), sc.transform(test[label].values.reshape(-1,1))
```

```python
def categorize(train,test,labels,target):
    df1,df2 = train.copy(), test.copy()
    idx,_ = df1.shape
    x = pd.concat([df1.drop(target,axis=1),df2],0)
    x = pd.get_dummies(x,columns=labels,drop_first=True)
    df1 = x.iloc[:idx]
    df2 = x.iloc[idx:]
    return df1,df2
```

```python
def rename_datetime(df1,df2):
    return df1.rename({'date_time':'ds'},inplace=True), df2.rename({'date_time':'ds'},inplace=True)
```

```python
def prepare_dataset(train, test, index, holiday, cont_cols,target):
    df1,df2 = train.copy(), test.copy()
    for col in cont_cols:
        df1[col],df2[col] = normalize(df1,df2,col)
    x = cont_cols+ cat_cols+ [index, target]
    holidays1 = df1[df1[holiday] != 'None'][[index,holiday]]
    holidays2 = df2[df2[holiday] != 'None'][[index,holiday]]
    holidays = pd.concat([holidays1,holidays2],0)
    for col in df1.columns:
        if col not in x:
            df1 = df1.drop(col,1)
            df2 = df2.drop(col,1)
    df1,df2 = df1.rename({index:'ds'}), df2.rename({index:'ds'})
    return df1,df2,holidays
```
